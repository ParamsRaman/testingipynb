# algo1: Vanilla k-means

Below we implement the traditional k-means algorithm (to use as baseline for evaluating our distributed implementations).

# algo2: Distributed kmeans (token passing version)

Algorithm to independently update the cluster centers on each machine (without requiring any conflicts). Data is partitioned into P machines and held fixed. Cluster centers move as tokens from one machine to another.

```python
import sys
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


def initCenters(points, K):
    mu = points.copy()
    np.random.shuffle(mu)
    return mu[:K]

def updatePoints(points, mu, mu_ind):
    distances = np.sqrt(((points - mu[:, np.newaxis])**2).sum(axis=2))
    return mu_ind[np.argmin(distances, axis=0)]

def updateCenters(points, zi, mu, mu_ind):
    ## if no points are assigned to a center, dont move the center
    ## (this check is to avoid nan s)
    for k in range(mu.shape[0]):
        if np.any(zi==mu_ind[k]) == False:
            print "potential nan found for k:%d, breaking.." % k
            return mu
    return np.array([points[zi==mu_ind[k]].mean(axis=0) for k in range(mu.shape[0])])

def hasConverged(oldmu, mu, t, norm_diff):
    #print "iteration: %d, norm difference: %f" % (t, np.linalg.norm(oldmu-mu))
    norm_diff[t, 0] = np.linalg.norm(oldmu-mu)
    #if np.allclose(oldmu, mu):
    #    print "Iteration: %d, converged in centers" % t
    #    return True
    if (t+1) == T:
        print "Iteration: %d, converged in max-iterations" % t
        return True
    return False

#######################
# N = data points     #
# D = dims            #
# K = components      #
# T = max iterations  #
#######################
N = 9000
D = 100
K = 20
T = 400
P = 4

print "Starting k-means with N=%d, D=%d, K=%d, P=%d " % (N, D, K, P)

#np.random.seed(11)
'''points = np.vstack(((np.random.randn(N/K, D) * 0.3 + np.array([6, 8])),
                  (np.random.randn(N/K, D) * 0.24 + np.array([-4, 6])),
                  (np.random.randn(N/K, D) * 0.45 + np.array([0, -4])),
                  (np.random.randn(N/K, D) * 0.3 + np.array([6, 0]))))'''
clusters = np.random.randn(K, D) * 0.8
points = np.zeros((K, N/K, D))
for k in range(0, K):
    #rand_sigma = 0.1 * np.random.random_sample()
    rand_sigma = 0.09
    #print "for k: %d, rand_sigma: %f" %(k, rand_sigma)
    points[k] = np.random.randn(N/K, D) * rand_sigma + clusters[k]
points = points.reshape(N, D)
np.random.shuffle(points)

## Assign data points to centers randomly
zi = np.random.randint(0, K, N)
oldzi = zi.copy()

oldmu = np.zeros((K, D))
#mu = 10 * np.random.sample((K, D)) - 5
mu = initCenters(points, K)

## store assignment of cluster centers to each machine
pk = np.zeros((P, 1))

norm_diff = np.zeros((T, 1))

t = 0
p = 0
debug = False
while not hasConverged(oldmu, mu, t, norm_diff):
    ## randomly sample cluster centers (w/o replacement)
    if (p == 0):
        pk = np.random.choice(K, K, replace=False)
        if debug == True:
            print "randomizing centers and generating pk:"
            print pk
        t = t+1
        
    oldmu = mu.copy()
    oldzi = zi.copy()
    
    ## to manage chunk of data points for machine p
    low = p*(N/P)
    high = p*(N/P)+(N/P)
    
    ## to manage chunk of cluster centers for machine p
    plow = p*(K/P)
    phigh = p*(K/P)+(K/P)
    mu_ind = pk[plow:phigh]
    
    if debug == True:
        print "------"
        print "p: %d" % p
        print "low: %d, high: %d" % (low, high)
        print "plow: %d, phigh: %d" % (plow, phigh)
        #print points[low:high, :]
        print mu_ind
        print oldmu[mu_ind]

    ## Assign points to closest centers in oldmu
    zi[np.where(np.in1d(zi,mu_ind))[0]] = updatePoints(points[np.where(np.in1d(zi,mu_ind))[0], :], oldmu[mu_ind], mu_ind)
    
    if debug == True:
        print zi[np.where(np.in1d(zi,mu_ind))[0]]
    
    ## Re-compute centers
    mu[mu_ind] = updateCenters(points[np.where(np.in1d(zi,mu_ind))[0], :], zi[np.where(np.in1d(zi,mu_ind))[0]], oldmu[mu_ind], mu_ind)
    
    if debug == True:
        print "mu[mu_ind]:"
        print mu[mu_ind]
    
    p = (p+1)%P
    

print "kmeans ended.."
plt.figure(0)
plt.scatter(points[:, 0], points[:, 1])
plt.scatter(mu[:, 0], mu[:, 1], c='r', s=100)
#plt.scatter(clusters[:, 0], clusters[:, 1], c='g', s=100)
#for i, txt in enumerate(range(0,K)):
#    plt.annotate(txt, (mu[i, 0], mu[i, 1]))
    
plt.figure(1)
plt.yscale('log')
plt.plot(range(0, T), norm_diff, c='m')    
```

# algo3: Distributed kmeans (each p runs in isolation)
